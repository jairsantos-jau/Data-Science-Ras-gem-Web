# Configuração de um raspador (scraper)

```python
##Configuração de um raspador (scraper)
import requests ##para trabalhar com web
import re ##regex
import json ##converte arquivos para json
from bs4 import BeautifulSoup ##raspador
```

```python
#pip install gtts
from gtts import gTTS ##para falar
import os ##trabalhar com o sistema
```

```python
result = requests.get('https://web.archive.org/web/20150320143424/http://allrecipes.com.br/receita/4630/arangini--bolinhos-de-arroz---italiana-.aspx?o_ln=SimRecipes_Link_2&o_is=SimilarRecipes')
```

```python
print(result.status_code) ##se tiver status code 200 a pagina está ok, se for 4O4, não encontrada, se for 403 o servidor não aceitou o raspador
```

```python
src = result.content
```

```python
soup = BeautifulSoup(src,'lxml')
soup
```

```python
soup.find('div')
```

```python
soup.find('div').find_all('span')
```

```python
nome_receita = soup.find('div').find_all("span",{'itemprop':'name'})
for a in nome_receita:
    nome_receita = a.text
nome_receita
```

```python
descricao = soup.find('div').find_all("span",{'itemprop':'description'})
for a in descricao:
    descricao = a.text
    print(descricao)
```

```python
ingredients = soup.find('div').find_all("span",{'itemprop':'ingredients'})
ingredients
```

```python
receita = soup.find('div').find_all("ol",{'itemprop':'recipeInstructions'})
for a in receita:
    receita = a.text
print(receita)
```

```python
ingredients_list = []
for a in ingredients:
    ingredients_list.append(a.text)
print(ingredients_list)
```

```python
#convertendo o script em dict
receita_dict = {'ID':1,
               'Titulo':nome_receita,
                'Desrição':descricao,
                'Ingredientes':ingredients_list,
                'Receita':receita}
receita_dict
```

```python
#convertendo para JSON
#ensure_ascii=False
receita_json = json.dumps(receita_dict,ensure_ascii=False)
print(receita_json)
```

```python
#em text informe o texto a ser dito evite usar acentuaçao
AudioDoTexto = gTTS(text = receita_dict['Receita'], lang='pt')
AudioDoTexto.save("C:/Users/FIC\Desktop/datasets/AULA03audio03.mp3")
```
